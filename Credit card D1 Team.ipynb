```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score,silhouette_score
from sklearn.model_selection import cross_val_score
```

```python
card = pd.read_csv("creditcard D1 team.csv")
```

```python
card.head()
```

```python
card.info()
```

```python
card['Time'].unique().tolist()
```

```python
train = card.copy()
```

```python
scaler = StandardScaler()
train['Time Encoded'] = scaler.fit_transform(train[['Time']])
train['Amount Encoded'] = scaler.fit_transform(train[['Amount']])
```

```python
train.head()
```

```python
train.drop(columns = ['Time', 'Amount'])
```

```python
X = train.drop(columns = 'Class')
isolation_forest = IsolationForest(n_estimators = 100, random_state = 0)
isolation_forest.fit(X)
```

```python
train['Anomaly'] = isolation_forest.predict(X)
```

```python
result = train[['Anomaly','Class']]
```

```python
result.head(20)
```

```python
train['Anomaly'].value_counts()
```

```python
dict1 = {1: 0, -1: 1}
train['CLASS'] = train['Anomaly'].replace(dict1)
```

```python
result1 = train[['Class','CLASS']]
result1
```

```python
Train = train[['CLASS']]
test = train[['Class']]
```

```python
ACC = accuracy_score(Train, test)
ACC
```
